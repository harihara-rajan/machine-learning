# Polynomial Regression
* Polynomial regression implementation using library 
* library used: scikit learn

## Importing the libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

## Importing the dataset

```python
dataset = pd.read_csv("polynomial_regression_data.csv")
X = dataset.iloc[:,1].values # position level
y = dataset.iloc[:,-1].values # salary
X = X.reshape(len(X),1)
y = y.reshape(len(y),1)
```

## Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() # creating new instance of Linear regression class 
regressor.fit(X,y) # fitting the model
```

## Training the Polynomial Regression model on the whole dataset
* We have only one feature in our data-set 
* Creating more features with the help of existing features is called as **Feature Mapping**
* $y = b_0 + b_1*x_1 + b_2*x_1^2 + ....+b_n*x_1^n$
* we perform this operation with the help of **PolynomialFeatures** class from **preprocessing** module

```python
from sklearn.preprocessing import PolynomialFeatures
x_mat = PolynomialFeatures(degree=3) # here 2 denotes the degree upto which feature mapping needs to be performed
X_poly = x_mat.fit_transform(X)
```

## Visualising the Linear Regression results
* Note that fitting a linear model to non-linear data results in **underfitting**
* To avoid this we have to choose a proper model, i.e. our hypthesis should have some **higher order polynomial terms** 

```python
y_pred = regressor.predict(X)
plt.scatter(X,y,c='r',marker='x',label='Data')
plt.plot(X,y_pred,c='k',label='Regression line')
plt.xlabel("Position Levels")
plt.ylabel("Salary")
plt.title("Position vs Salary")
plt.legend()
plt.show()
```

## Visualising the Polynomial Regression results
* Change the degree in PolynomialFeature class and visualise the plot
* As we choose higher degree, we will get a good fit 
* also note that if we choose **higher degree**, it will also lead to **overfitting problem**
* Careful selection of degree is required.In this case 3 seems to be a good fit 

```python
poly_regressor = LinearRegression() # creating new instance of Linear regression class 
poly_regressor.fit(X_poly,y) # fitting the model
y_poly_pred = poly_regressor.predict(X_poly)
plt.scatter(X,y,c='r',marker='x',label='Data')
plt.plot(X,y_poly_pred,c='k',label='Regression line')
plt.xlabel("Position Levels")
plt.ylabel("Salary")
plt.title("Position vs Salary")
plt.legend()
plt.show()
```

## Predicting a new result with Linear Regression

```python
regressor.predict([[6.5]])
```

## Predicting a new result with Polynomial Regression

```python
poly_regressor.predict(x_mat.fit_transform([[6.5]]))
```
